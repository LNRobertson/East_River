Create a separate dataframe for the load control columns with missing data to run KNN imputation. Keeping the original dataframe intact in case we are able to procure actual load control data. 

````python
from sklearn.impute import KNNImputer

# Define the target columns (note: using 'Control_Threshold_MW')
load_cols = ['Load_Control_KW', 'Load_Control_MW', 'Estimated_Online_Load_Control_MW', 'Control_Threshold_MW']

# Create a separate dataframe for these columns
df_load_controls = df_weather_scada[load_cols].copy()

# Impute missing values using KNNImputer
imputer = KNNImputer(n_neighbors=5)
df_load_controls_imputed = pd.DataFrame(imputer.fit_transform(df_load_controls), columns=load_cols)

# View the head of the imputed dataframe for confirmation
print(df_load_controls_imputed.head())
````

````python
# Create a boolean mask of where the original dataframe had missing values
imputed_mask = df_load_controls.isna()

# Count the number of missing (and therefore imputed) values per column
imputed_counts = imputed_mask.sum()
print("Number of imputed values per column:")
print(imputed_counts)

# Optionally, view a small sample of the imputed values by selecting
# the imputed entries from the imputed dataframe.
imputed_values_sample = df_load_controls_imputed.where(imputed_mask)
print("\nSample of imputed values:")
print(imputed_values_sample.head(10))
````

````python
# Count the number of imputed values per column:
imputed_values_count = (df_load_controls.isna() & df_load_controls_imputed.notna()).sum()
print("Imputed values count per column:")
print(imputed_values_count)

# Verify these counts match the number of missing values in the original dataframe:
original_missing = df_load_controls.isna().sum()
print("\nMissing values in the original dataframe:")
print(original_missing)

# Assert that the imputation filled exactly the missing values:
assert (imputed_values_count == original_missing).all(), "Mismatch in counts!"
print("\nVerification successful: All missing values were imputed correctly.")
````

````python
df_weather_scada[load_cols] = df_load_controls_imputed
````

````python
# For example, let's check the dates in the first week of January 2021
date_range = (df_weather_scada['datetime'] >= '2021-01-01') & (df_weather_scada['datetime'] < '2021-01-08')
jan_sample = df_weather_scada[date_range]

# Identify any rows where the day_of_week does not match the day name computed from datetime
mismatches = jan_sample[jan_sample['day_of_week'] != jan_sample['datetime'].dt.day_name()]

print("Found {} mismatches in day names".format(len(mismatches)))
print(mismatches[['datetime', 'day_of_week']])
````

````python
df_wx_history = pd.read_csv(r"C:\Users\Linds\Repos\East_River\data\Raw\DSU_wx_history.csv")
print(df_wx_history.head())
````

````python
# Ensure that df_wx_history's obs_tms_lcl is converted to datetime with UTC
df_wx_history['datetime'] = pd.to_datetime(df_wx_history['obs_tms_lcl'], format='%Y-%m-%d %H:%M:%S.%f', errors='coerce')
df_wx_history['datetime'] = df_wx_history['datetime'].dt.tz_localize('UTC')

# Perform an outer merge on location and datetime to see mismatches
comparison = pd.merge(
    df_weather_scada[['location', 'datetime']],
    df_wx_history[['location', 'datetime']],
    on=['location', 'datetime'],
    how='outer',
    indicator=True
)

print("Rows present only in df_weather_scada:")
print(comparison[comparison['_merge'] == 'left_only'])

print("\nRows present only in df_wx_history:")
print(comparison[comparison['_merge'] == 'right_only'])

print("\nMerge indicator summary:")
print(comparison['_merge'].value_counts())
````

````python
# Identify common columns between df_weather_scada and df_wx_history
common_cols = df_weather_scada.columns.intersection(df_wx_history.columns)
print("Common columns:")
print(common_cols)

# Optionally, create subsets using only the common columns
df_weather_subset = df_weather_scada[common_cols]
df_wx_subset = df_wx_history[common_cols]

# Now, if you're merging the two datasets (or comparing them), you can use the common columns.
merged_df = pd.merge(df_weather_subset, df_wx_subset, on=['location', 'datetime'], how='outer', indicator=True)
print(merged_df.head())
````

````python
import holidays

# Generate US holidays covering the years in our datetime column
years = range(df_weather_scada['datetime'].dt.year.min(), df_weather_scada['datetime'].dt.year.max() + 1)
us_holidays = holidays.US(years=years)

# Create a boolean 'is_holiday' flag: True if the date is a US holiday, False otherwise.
df_weather_scada['is_holiday'] = df_weather_scada['datetime'].dt.date.apply(lambda x: x in us_holidays)

# Display a sample to verify
print(df_weather_scada[['datetime', 'is_holiday']].head(10))
````
